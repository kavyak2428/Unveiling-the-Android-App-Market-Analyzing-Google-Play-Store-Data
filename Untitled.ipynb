```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud
from textblob import TextBlob
```

```python
# Load datasets
apps_df = pd.read_csv("apps.csv")
reviews_df = pd.read_csv("user_reviews.csv")
```

```python
# Data Cleaning
apps_df.dropna(inplace=True)  # Removing missing values
reviews_df.dropna(inplace=True)
```

```python
# Display basic information
print("Apps Data Overview:")
print(apps_df.info())
print("\nUser Reviews Data Overview:")
print(reviews_df.info())
```

```python
# Convert columns to appropriate data types
apps_df["Installs"] = apps_df["Installs"].str.replace(",", "").str.replace("+", "").astype(int)
apps_df["Price"] = apps_df["Price"].str.replace("$", "").astype(float)
```

```python
# Category Analysis
category_counts = apps_df["Category"].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(y=category_counts.index, x=category_counts.values, palette="viridis")
plt.title("App Distribution Across Categories")
plt.xlabel("Number of Apps")
plt.ylabel("Category")
plt.show()
```

```python
# Ratings Analysis
plt.figure(figsize=(8, 5))
sns.histplot(apps_df["Rating"].dropna(), bins=30, kde=True, color='blue')
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Sentiment Analysis
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
reviews_df["Sentiment Score"] = reviews_df["Translated_Review"].dropna().apply(get_sentiment)
plt.figure(figsize=(8, 5))
sns.histplot(reviews_df["Sentiment Score"], bins=30, kde=True, color='green')
plt.title("Sentiment Score Distribution")
plt.xlabel("Sentiment Score")
plt.ylabel("Count")
plt.show()
```

```python
# WordCloud for Common Words in Reviews
text = " ".join(review for review in reviews_df["Translated_Review"].dropna())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Common Words in Reviews")
plt.show()
```
